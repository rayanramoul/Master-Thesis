# How To Use Spectral Clustering with Sickit Learn


## Method Used

In practice Spectral Clustering is useful when the structure of the individual clusters is highly non-convex or more generally when a measure of the center and spread of the cluster is not a suitable description of the complete cluster. For instance when clusters are nested circles on the 2D plane.

## Generate Data

```python
from sklearn.datasets import make_circles

# generate a synthetic circle dataset
data, labels = make_circles(n_samples=500, factor=0.5, random_state=0, noise = 0.05)
```

## Visualize Data

```python
plt.scatter(data[:,0],data[:,1],c=labels)
```

## Creating the Clusters

```python
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
spectralC = SpectralClustering(n_clusters=2, eigen_solver='arpack', affinity="nearest_neighbors")
spectralC.fit(data)
kmeansC = KMeans(n_clusters=2)
kmeansC.fit(data)
```

```python
spectralC.labels_
```

```python
spectralC
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(8,6))
ax1.set_title('Spectral Clustering')
ax1.scatter(data[:,0],data[:,1],c=spectralC.labels_)
ax2.set_title("K-means")
ax2.scatter(data[:,0],data[:,1],c=kmeansC.labels_)
```
