# Clustering with Python and scikit-learn


## Method Used

### K Means

k-Means Clustering clusters data based on their distance to the means of clusters.

Given a set *X of n* points in a *d-dimensional* space and an integer *k* that represents the number of clusters, the algorithm iterates through the following steps:

1. Randomly pick *k** cluster centers {c1,…,ck} in the d-dimensional space in order to form clusters {C1, C2,…,Ck}

2. For each *i**, set the cluster Ci to be the set of points in X that are closer to ci than they are to cj for all i≠j

3. For each *i** let ci be the center of cluster Ci (mean of the vectors in Ci)

4. Repeat until convergence


### Hierachical Clustering

Recursively merges the pair of clusters that minimally increases a given linkage distance.
Given D =(*Dij*) a matrix of dissimilarities between data, the algorithm iterates through the following steps:

1. Initially, clusters (groups) are singletons containing data

2. Find r and s groups such as Drs = min_ij (*Dij*)

3. Merge groups r and s into a group t

4. Build a new D matrix in which r and s disappeared and the new t group has distance from each group i, *Dti*: to be defined

5. Repeat the process until you find a single group


## Import Libraries

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.cluster import ward_tree
import numpy as np
```

## Generate Data

```python
# generate a synthetic circle dataset
data, labels = make_circles(n_samples=500, factor=0.5, random_state=0, noise = 0.05)
```

## Visualize Data

```python
plt.scatter(data[:,0],data[:,1],c=labels)
```

## Creating the Clusters

### K Means

```python
Ckmeans = KMeans(n_clusters=2)
```

```python
Ckmeans.fit(data)
```

```python
Ckmeans.cluster_centers_
```

```python
Ckmeans.labels_
```

### Hierarchical Clustering

```python
Chclust = AgglomerativeClustering(linkage='complete').fit(data)
```

```python
Chclust.labels_
```

```python
Chclust
```

```python
Chclust.labels_
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(8,6))
ax1.set_title('K Means')
ax1.scatter(data[:,0],data[:,1],c=Ckmeans.labels_)
ax2.set_title("Original")
ax2.scatter(data[:,0],data[:,1],c=labels)
ax3.set_title('Hierarchical Clustering')
ax3.scatter(data[:,0],data[:,1],c=Chclust.labels_)
```

# The Use of Spicy package

```python
# Dendrogram visualization using Spicy
from scipy import cluster
from scipy.cluster import hierarchy
CChclust = hierarchy.linkage(data, 'single')
plt.figure()
dn = hierarchy.dendrogram(CChclust)
cutree = cluster.hierarchy.cut_tree(CChclust, n_clusters=2)
```
