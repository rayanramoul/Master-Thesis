# K Means Clustering with Python and scikit-learn


## Method Used

k Means Clustering clusters data based on their distance. 

Given a set *X* of *n* points in a *d*-dimensional space and an integer *k* that represents the number of clusters, the algorithm iterates through the following steps:

*   Randomly pick *k* cluster centers *{c1,…,ck}* in the *d*-dimensional space in order to form clusters {C1, C2,…,Ck}

*   For each *i*, set the cluster Ci to be the set of points in *X* that are closer to ci than they are to *cj* for all i≠j

*   For each i let ci be the center of cluster Ci (mean of the vectors in Ci)

*   Repeat until convergence


## Import Libraries

```python

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

## Generate Data

```python
data = make_blobs(n_samples=400, n_features=2, 
                           centers=3, cluster_std=1.2)
```

## Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

## Creating the Clusters

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(8,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```
